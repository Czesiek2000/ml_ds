import matplotlib and seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

set inline displaying matplotlib images

```python
%matplotlib inline
```

```python
data = sns.load_dataset("titanic")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
# drop nan values in each column
data = data.dropna()
```

line plot - Show trends over a sequence (time, age groups, etc.).

```python
# count number of people in each age group
age_counts = data['age'].value_counts()

# sort by age
age_counts = age_counts.sort_index()
```

```python
# plot age and number of people with that age
plt.figure(figsize=(10, 6))
plt.plot('age', 'age', data=data)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()


```

```python
# plot with markers
plt.figure(figsize=(10, 6))
plt.plot('age', 'age', data=data, marker='o')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```

```python
# plot age and number of people with that age with sns
sns.lineplot(x='age', y='age', data=data)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```

```python
# scatter chart - age vs fare
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['fare'])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=data)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
